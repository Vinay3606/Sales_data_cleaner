```python
# Install the python libraries
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Main Project ds/task 2/sales.csv')
```

```python
usd_to_inr=83
```

```python
df
```

# Clean The Data

```python
df['product']=df['product'].astype(str).str.replace('"','').str.strip()
df
```

```python
df['price']=df['price'].astype(str).str.replace("$",'').astype(float)
df
```

# Remove Duplicate

```python
df.drop_duplicates(subset=['product','price'],inplace=True)
```

```python
df
```

# Convert USD into INR

```python
df['price_inr']=df['price'] * usd_to_inr
df
```

```python
df=df.drop(columns=['price'])
```

```python
df
```

# Clean Data Into Json File

```python
df_json=df.to_json(orient='records',indent=4)
```

```python
print(df_json)
```

```python
save_path='/content/drive/MyDrive/Main Project ds/task 2/clean_file.json'
```

```python
with open(save_path,'w') as file:
  file.write(df_json)
```
